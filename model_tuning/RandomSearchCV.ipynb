## Model Tuning using RandomSearchCV

```python
# importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('species',axis = 1)
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

### for SVC

```python
from sklearn.svm import SVC
model_svm = SVC(C = 10, kernel='linear', gamma= 'auto')
model_svm.fit(X_train,y_train)
```

```python
model_svm.score(X_test,y_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV
     
classifier = RandomizedSearchCV((model_svm),{ # we put the multiples hyperparameters in svm model
    'C' : [1,10,20,30],
    'kernel' :['rbf','linear'],
},n_iter=4 ,cv = 5,return_train_score = False)
     
classifier.fit(X,y)
```

```python
results = pd.DataFrame(classifier.cv_results_)
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```

### Hence during the hyperparameter tuning (using RandomizedsearchCV) in SVM model the best accuracy comes from putting the parameters to C=10,kernal=rbf.
