## Netflix Data Analysis

```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset
df = pd.read_csv("netflix_titles.csv")
```

```python
# displaying the first few rows of the dataset
df.head()
```

```python
# displaying the summary of the dataset
df.info()
```

```python
# checking for missing values
df.isnull().sum()
```

```python
# dropping rows with missing values
df.dropna(inplace=True)
```

```python
# dropping duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# displaying the count of each type of content (Movie or TV Show)
df['type'].value_counts()
```

```python
# displaying the count of content produced in each country
df['country'].value_counts().head(10)
```

```python
# visualizing the count of each type of content
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows")
plt.show()
```

```python
# visualizing the count of content produced in each country

top_country = df['country'].value_counts().head(10)
top_country.plot(kind='bar')

plt.title("Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
# displaying the columns of the dataset
df.columns
```

```python
# analyzing the release year of the content
year_count = df['release_year'].value_counts()
```

```python
# sorting the release year count in ascending order and displaying it
year_count = year_count.sort_index()
print(year_count)
```

```python
# visualizing the release year count

plt.figure(figsize=(12,6))
year_count.plot(kind='line')
plt.title("Netflix Content Released Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Shows")
plt.show()
```

### Conclusion
#### The analysis of the Netflix dataset shows that Netflix contains significantly more movies than TV shows, indicating that the platform mainly focuses on movie content. The data also reveals that the United States produces the highest amount of Netflix content compared to other countries. In addition, the number of movies and TV shows released on Netflix increased rapidly after 2015, showing the platform’s major growth and expansion in recent years.
